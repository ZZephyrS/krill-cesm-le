```python
ds = xe.util.grid_global(1, 1)

lon=ds.lon
lat=ds.lat

a = (lat <= -45) & (lon > 270) | (lon <= 0)
b = (lat <= -45) & (lon > 270) | (lon <= 25)

c = (lat <= -45) & (lon > 260) | (lon <= 25)
d = (lat <= -45) & (lon > 270) | (lon <= 0)

#f, ax = plot.subplots(proj='spstere')
#ax.pcolormesh(lon, lat, test, cmap='green')
#ax.format(boundinglat=-45, land=True)
```

```python
f, axs = plot.subplots(ncols=5, proj = 'splaea')
axs[0].pcolormesh(lon, lat, a, cmap='green')
axs[1].pcolormesh(lon, lat, b, cmap='skyblue')
axs[2].pcolormesh(lon, lat, c, cmap='red')
axs[3].pcolormesh(lon, lat, d, cmap='purple')
axs[4].pcolormesh(lon, lat, e, cmap='grey')
axs.format(land = True, boundinglat=-40)
```
